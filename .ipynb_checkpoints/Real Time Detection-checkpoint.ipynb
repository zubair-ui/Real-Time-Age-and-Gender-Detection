# Real Time Detection using the Fine-tuned Mobilenet-V2 Model


## Setup

```python
import cv2
import numpy as np
from tensorflow.keras.models import load_model

model = load_model("saved_model/age_gender_model_final.keras")

IMG_SIZE = 128 
```

## Real-Time Video Capture

```python
cap = cv2.VideoCapture(0)

while True:
    ret, frame = cap.read()
    if not ret:
        break

    img = cv2.resize(frame, (IMG_SIZE, IMG_SIZE))
    img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    img_norm = img_rgb / 255.0
    img_input = np.expand_dims(img_norm, axis=0)

    age_pred, gender_pred = model.predict(img_input, verbose=0)
    age = int(age_pred[0][0])
    gender = "Male" if gender_pred[0][0] > 0.5 else "Female"

    cv2.putText(frame, f"Age: {age}", (10, 30),
                cv2.FONT_HERSHEY_SIMPLEX, 1, (0, 255, 0), 2)
    cv2.putText(frame, f"Gender: {gender}", (10, 70),
                cv2.FONT_HERSHEY_SIMPLEX, 1, (0, 255, 255), 2)

    cv2.imshow("Age & Gender Detection", frame)

    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

cap.release()
cv2.destroyAllWindows()
```
